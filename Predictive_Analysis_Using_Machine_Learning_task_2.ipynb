```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression
from pyspark.ml.evaluation import RegressionEvaluator
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("PredictiveAnalysis") \
    .getOrCreate()
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = spark.read.csv("sales_data_sample.csv", header=True, inferSchema=True)
```

```python
Dataset Loaded
```

```python
df.show(5)
```

```python
Feature Selection
```

```python

data = df.select("QUANTITYORDERED", "PRICEEACH", "SALES")
data.show(5)
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression
from pyspark.ml.evaluation import RegressionEvaluator
```

```python
Feature Vector Creation
```

```python
assembler = VectorAssembler(
    inputCols=["QUANTITYORDERED", "PRICEEACH"],
    outputCol="features"
)

final_data = assembler.transform(data)
final_data.show(5)
```

```python
train_data, test_data = final_data.randomSplit([0.8, 0.2], seed=42)
```

```python
lr = LinearRegression(featuresCol="features", labelCol="SALES")

model = lr.fit(train_data)
```

```python
Predictions
```

```python
predictions = model.transform(test_data)

predictions.select("SALES", "prediction").show(5)
```

```python
RMSE Result
```

```python
evaluator = RegressionEvaluator(
    labelCol="SALES",
    predictionCol="prediction",
    metricName="rmse"
)

rmse = evaluator.evaluate(predictions)

print("RMSE:", rmse)
```

```python
## Result

The Linear Regression model was successfully trained using PySpark MLlib to predict sales values based on quantity ordered and product price. Predictions were generated on the test dataset, and the model achieved an RMSE value of approximately 1011.86.
```
